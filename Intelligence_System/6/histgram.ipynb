```python
import numpy as np
import matplotlib.pyplot as plt

def kai_distribution (n):
    x = np.zeros(1000)
    for i in range(n):
        x = x + [k ** 2 for k in np.random.normal(0, 1, 1000)]
    plt.figure()
    plt.xlim(0,30)
    plt.ylim(0,500)
    plt.hist(x, range=(0, 30),bins=80)
    #filename="hist" + str(i+1) + ".png"
    plt.title("hist" + str(i+1))
    #plt.savefig(filename)
    plt.show()
```

```python
kai_distribution(1)
kai_distribution(2)
kai_distribution(3)
kai_distribution(4)
kai_distribution(5)
```
