```python
#use method of multipliers 
#dual ascent method is a case that c = 0
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x_l = []
y_l = []
lam_l = []
def f (x,y):
    return 3*(x**2)+2*(y**2)
def multipliers (x,y,lam,c,n):
    x_l.append(x)
    y_l.append(y)
    lam_l.append(lam)
    for i in range(n):
        x = (c-lam)/(6+5*c/2)
        y = (c-lam)/(4+5*c/3)
        #iとlam, i と(x,y)をプロット
        lam = lam + c * (x + y - 1) 
        x_l.append(x)
        y_l.append(y)
        lam_l.append(lam)
    fig, (axL, axR) = plt.subplots(ncols=2, figsize=(8,4))

    axL.set_xlim([-0.2, 1])
    axL.set_ylim([-0.2, 1])
    X,Y=np.mgrid[-0.1:1:0.01,-0.2:1:0.01]
    Z = f(X,Y)
    x1 = np.arange(-0.1, 1,0.1)
    y1 = -x1+1
    axL.plot(x1, y1, color = "blue")
    axL.contour(X,Y,Z,levels=[0.3, 0.6, 0.9, 1.2, 1.5,1.8])
    for i in (range(len(x_l))):
        axL.annotate(i,xy=(x_l[i],y_l[i]))
    axL.plot(x_l,y_l)
    axL.set_xlabel('x')
    axL.set_ylabel('y')

    axR.plot(lam_l, linewidth=2)
    axR.set_xlabel('Iteration')
    axR.set_ylabel('lambda')

    fig.show()
    plt.savefig('multipliers.png')
```

```python
multipliers(0,0,0,5,5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x_l = []
y_l = []
lam_l = []
def f (x,y):
    return 3*(x**2)+2*(y**2)
def dual_ascent (x,y,lam,epsilon,n):
    x_l.append(x)
    y_l.append(y)
    lam_l.append(lam)
    for i in range(n):
        x = -lam/6
        y = -lam/4
        lam = lam + epsilon * (x + y - 1)
        x_l.append(x)
        y_l.append(y)
        lam_l.append(lam)
    
    fig, (axL, axR) = plt.subplots(ncols=2, figsize=(8,4))

    axL.set_xlim([-0.2, 1])
    axL.set_ylim([-0.2, 1])
    X,Y=np.mgrid[-0.1:1:0.01,-0.2:1:0.01]
    Z = f(X,Y)
    x1 = np.arange(-0.1, 1,0.1)
    y1 = -x1+1
    axL.plot(x1, y1, color = "blue")
    axL.contour(X,Y,Z,levels=[0.3, 0.6, 0.9, 1.2, 1.5])
    for i in (range(len(x_l))):
        axL.annotate(i,xy=(x_l[i],y_l[i]))
    axL.plot(x_l,y_l)
    axL.set_xlabel('x')
    axL.set_ylabel('y')

    axR.plot(lam_l, linewidth=2)
    axR.set_xlabel('Iteration')
    axR.set_ylabel('lambda')

    fig.show()
    plt.savefig('dual_ascent.png')
```

```python
dual_ascent(0,0,0,2,5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x_l = []
y_l = []
mu_l = []
def f (x,y):
    return 3*(x**2)+2*(y**2)
def gradient_method (x,y,mu,epsilon,c):
    x_l.append(x)
    y_l.append(y)
    mu_l.append(mu)
    for i in range(c):
        x = mu/6
        y = mu/4
        mu = max(0,mu+epsilon*(-5*mu/12+1))
        x_l.append(x)
        y_l.append(y)
        mu_l.append(mu)
        
    fig, (axL, axR) = plt.subplots(ncols=2, figsize=(8,4))

    axL.set_xlim([-0.6, 1])
    axL.set_ylim([-0.6, 1])
    X,Y=np.mgrid[-0.6:1:0.01,-0.6:1:0.01]
    Z = f(X,Y)
    x1 = np.arange(-0.1, 1,0.1)
    y1 = -x1+1
    axL.plot(x1, y1, color = "blue")
    axL.contour(X,Y,Z,levels=[0.3, 0.6, 0.9, 1.2, 1.5])
    for i in (range(len(x_l))):
        axL.annotate(i,xy=(x_l[i],y_l[i]))
    axL.plot(x_l,y_l)
    axL.set_xlabel('x')
    axL.set_ylabel('y')

    axR.plot(mu_l, linewidth=2)
    axR.set_xlabel('Iteration')
    axR.set_ylabel('mu')

    fig.show()
    plt.savefig('gradient.png')
```
