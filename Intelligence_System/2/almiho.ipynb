```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def f(x1, x2):
    return 10*(x1**2) + x2**2
```

```python
def back_track (x1,x2,alpha,beta):
    epsilon = 1
    while(f(x1-epsilon*20*x1,x2-2*epsilon*x2)-f(x1,x2) > -alpha*epsilon*(400*(x1**2)+4*(x2**2))):
        epsilon = beta * epsilon
    return epsilon
```

```python
c = 0
x1 = 5
x2 = 17
alpha = 0.5
beta = 0.8
x1_l = [x1]
x2_l = [x2]
while(c<=5):
    e = back_track(x1,x2,alpha,beta)
    x1 = x1 - e*20*x1
    x1_l.append(x1)
    x2 = x2 - e*2*x2
    x2_l.append(x2)
    c=c+1
```

```python
X, Y = np.mgrid[-10:11, -20:21]
Z = f(X,Y)
plt.figure()
fig, ax = plt.subplots(figsize=(3, 6))
ax.contour(X, Y, Z,)
plt.plot(x1_l, x2_l)
plt.savefig('map.png')
plt.show()
```
