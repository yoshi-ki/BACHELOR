```python
#値を求めるプログラム
def newton_method (c,n):
    if(abs(c*c*c - 1) < 0.01):
        return c
    else :
        c = 2*c/3 + 1/(3*c*c)
        return newton_method(c,n+1)
```

```python
for i in range(100):
    for j in range(100):
        a = 1 - 0.02*i
        b = 1 - 0.02*j
        if(a!=0 and b!=0):
            print(newton_method(complex(a,b),0))
```

```python
import numpy as np
#回数をグラフにするプログラム
def newton_method (c,n):
    if(abs(c*c*c - 1) < 0.01):
        return n
    else :
        c = 2*c/3 + 1/(3*c*c)
        return newton_method(c,n+1)
x = np.array([])
y = np.array([])
z = np.array([])
for i in range(100):
    for j in range(100):
        a = -1 + 0.02*i
        x = np.append(x,a)
        b = -1 + 0.02*j
        y = np.append(y,b)
        if(a!=0 and b!=0):
            z = np.append(z,newton_method(complex(a,b),0))
        else:
            z = np.append(z,0)
```

```python
#描画
plt.figure()
plt.scatter(x, y, s=100, c=z, cmap='Blues')
plt.savefig('heatmap.png')
```
